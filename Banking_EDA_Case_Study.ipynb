```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#Genarating descriptive statustics for the data frame
df.describe()
```

```python
bins =[0,100000,300000,float('inf')]
labels=['Low','Medium','high']

df['Income Band']= pd.cut(df['Estimated Income'],bins=bins,labels=labels,right=False)

df['Income Band'].value_counts().plot(kind='bar')
```

```python
#Examine the destribution of unique catogories in categorical columns

categorical_cols= df[['BRId','GenderId','IAId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','Income Band']].columns
for col in categorical_cols:
  print(f"Value counts for '{col}' :")
  display(df[col].value_counts())
```


Univariate Analysis



```python
for i, predictor in enumerate(df[['BRId','GenderId','IAId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','Income Band']].columns):
  plt.figure(i)
  sns.countplot(data=df,x=predictor)
```

Bivariate Analysis

```python
for i, predictor in enumerate(df[['BRId','GenderId','IAId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','Income Band']].columns):
  plt.figure(i)
  sns.countplot(data=df,x=predictor,hue='Nationality')
```

```python
#Histogram of value_counts for diffrent Occupation

for col in categorical_cols:
  if col == 'Occupation':
    continue
  plt.figure(figsize=(6,4))
  sns.histplot(df[col])
  plt.title('Histogram of Occoupation Count')
  plt.xlabel(col)
  plt.ylabel('Count')
  plt.show()


```

Numarical Analysis

```python
numarical_cols =['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']
#Univarient Analysis
plt.figure(figsize=(15,15))
for i,col in enumerate(numarical_cols):
  plt.subplot(4,3,i+1)
  sns.histplot(df[col],kde=True)
  plt.title(f'Histogram of {col}')
  plt.tight_layout()
plt.show()
```

Heatmaps

```python
numarical_cols =['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']
corelation_matrix= df[numarical_cols].corr()

plt.figure(figsize=(12,12))
sns.heatmap(corelation_matrix,annot=True,cmap='crest',fmt = '.2f')
plt.title('Correlation Heatmap')
plt.show()
```

## Insights of EDA:

The strongest positive correlation occurs among "Bank Deposits" with "Checking Accounts" and "Foreign Currency Account" indicating that customers who maintain hign balances in one account type often hold substantial amount/funds across other accounts as well
